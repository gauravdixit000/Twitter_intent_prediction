```python
import nltk
from nltk.corpus import stopwords
import string
```

# Remove stopwords and Punctuation.

```python
stop_list = stopwords.words('english')
punc = string.punctuation
```

```python
def clean(strg):
    list1 = []
    list2 = strg.split()
    for i in list2:
        if i in stop_list:
            pass
        else:
            list1.append(i.lower())
    strg1 = " ".join(list1)
    list3 = []
    for i in strg1:
        if i in punc:
            pass
        else:
            list3.append(i)
    strg2 = "".join(list3)
    
    return strg2
```

```python
train_data['clean_text'] = train_data['tweet_text'].apply(clean)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
x_train,x_val,y_train,y_val = train_test_split(df1,Y,test_size=0.25,random_state=50,shuffle=True)
```

```python
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
from datetime import datetime
def perform_model(model,x_train, y_train, x_val, y_val, cm_normalize=True, \
                 print_cm=True, cm_cmap=plt.cm.Greens):
    
    # to store results at various phrase
    results = dict()
    
    #time at which model starts training
    train_start_time = datetime.now()
    print('training the model..')
    model.fit(x_train, y_train)
    print('Done \n \n')
    train_end_time = datetime.now()
    results['training_time'] = train_end_time - train_start_time
    print('training_time(HH:MM:SS.ms) - {}\n\n'.format(results['training_time']))
    
    #predict train data
    print('Predicting train data')
    train_start_time = datetime.now()
    y_pred_train = model.predict(x_train)
    train_end_time = datetime.now()
    print('Done \n \n')
    results['testing_train_time'] = train_start_time - train_end_time
    print('testing_time(HH:MM:SS.ms) - {}\n\n'.format(results['testing_train_time']))
    results['predicted_train'] = y_pred_train
    
    #predict test data
    print('Predicting test data')
    test_start_time = datetime.now()
    y_pred_test = model.predict(x_val)
    test_end_time = datetime.now()
    print('Done \n \n')
    results['testing_test_time'] = test_end_time - test_start_time
    print('testing_time(HH:MM:SS.ms) - {}\n\n'.format(results['testing_test_time']))
    results['predicted_test'] = y_pred_test
    
    #calculating accuracy of the model on train data
    accuracy_train = metrics.accuracy_score(y_true=y_train,y_pred=y_pred_train)
    # store accuracy in results
    results['accuracy_train'] = accuracy_train
    print('-------------------------------------')
    print('         Accuracy On Train Data      ')
    print('-------------------------------------')
    print('\n       {}\n\n'.format(accuracy_train))
    
    
    #calculating accuracy of the model on test data
    accuracy_test = metrics.accuracy_score(y_true=y_val,y_pred=y_pred_test)
    # store accuracy in results
    results['accuracy_test'] = accuracy_test
    print('-----------------------------------')
    print('         Accuracy On Test Data     ')
    print('-----------------------------------')
    print('\n       {}\n\n'.format(accuracy_test))
    
    
    #confusion matrix
    cm = metrics.confusion_matrix(y_val,y_pred_test)
    results['confusion_matrix'] = cm
    if print_cm:
        print('---------------------------')
        print('      Confusion Matrix     ')
        print('---------------------------')
        print('\n{}'.format(cm))
        
        
        
    # plot confusion matrix
    plt.figure(figsize=(10,8))
    sns.heatmap(cm,annot=True,fmt='.2f',square=True)
    plt.show()
    
    
    #get classification report
    print('----------------------')
    print('  Classificaton Report ')
    print('----------------------')
    classification_report = metrics.classification_report(y_val,y_pred_test)
    # store report in results
    results['Classification_report'] = classification_report
    print(classification_report)
    
    # add the trained model to the results
    results['model'] = model
     
    return results
    
    
    
```
